```python
import geomappy as mp
```

```python
nc = mp.Map("data/dis_2019_monthlymeans_cropped_complete.nc", epsg=3035)
```

```python
nc.plot_map(basemap=True, xticks=5, yticks=5, cmap="Blues", bins=[0,0.01,0.1,1,2,5,10,20,50,100,200,500],
            resolution='10m')
```

```python
geoms, values = nc.cross_profile((10, 50), (15, 60))
```

```python
geoms.plot_shapes(basemap=True, values=values, xticks=2, yticks=2, resolution='10m', 
                  cmap="Blues", bins=[0,0.01,0.1,1,2,5,10,20,50,100,200,500])
```
