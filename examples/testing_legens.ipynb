#### Legends
Legend are object based and can be created outside the plotting functions. This gives us the opportunity to test them before plotting them on the final maps

```python
import os
os.chdir("..")
```

```python
import geomappy as mp
import matplotlib.pyplot as plt
```

```python
water_table_depth = mp.Raster("data/wtd.tif")
```

Internally the plotting function uses either the ``LegendPatches`` or ``Colorbar`` object to draw the legend on the plot. To have more control on how they are drawn they can be passed in directly. The both have an ``example`` method that shows how it will look like in the plot (if the plotting takes too long to do it over and over again)

```python
l = mp.LegendPatches()
l.example()
l = mp.Colorbar()
l.example()
```

The ``example`` method also accepts a basemap, to make life easier

```python
ax = mp.basemap()
l = mp.LegendPatches()
l.example(ax=ax)
ax = mp.basemap()
c = mp.Colorbar()
c.example(ax=ax)
```

```python
# Changing some settings in the Colorbar object
ax = mp.basemap()
c = mp.Colorbar(title='Title', label='Label', label_font={'rotation':270, 'labelpad':20}, title_font={'fontsize':15})
c.example(ax=ax)
```

To see the Colorbar presented on a real plot:

```python
ax = mp.basemap()
water_table_depth.plot_map(basemap=True, bins=[0,0.1,0.5,1,2,5,10,25], cmap="Blues_r", ax=ax, 
                           legend=c, bounds='global')
```
