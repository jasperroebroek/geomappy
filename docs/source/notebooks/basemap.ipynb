# Creating a standalone basemap

The basemap functionality of `geomappy` is a convenience function build on top of `cartopy`, inspired on the old Basemap package.

```python
import geomappy as mp
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
```

Creating the most straightforward global map based on the PlateCarree projection

```python
ax = mp.basemap()
ax.coastlines()
plt.show()
```

Adding gridlines every 30 degrees

```python
ax = mp.basemap()
ax.coastlines()
mp.add_gridlines(ax, 30)
plt.show()
```

Adding labels

```python
ax = mp.basemap()
ax.coastlines()
mp.add_gridlines(ax, 30)
mp.add_ticks(ax, 30)
plt.show()
```

Changing font size

```python
ax = mp.basemap()
ax.coastlines()
mp.add_gridlines(ax, 30)
mp.add_ticks(ax, 30, fontsize=8)
plt.show()
```

Different projections can be either passed as projection or epsg code

```python
ax = mp.basemap(epsg=3035)
ax.coastlines()
mp.add_gridlines(ax, 15)

# differentiate between top and other labels
g = mp.add_ticks(ax, 60, fontsize=8)
g.top_labels = True

mp.add_ticks(ax, 15, fontsize=8)
g.right_labels = True

plt.show()
```

```python
ax = mp.basemap(projection=ccrs.Mercator())
ax.coastlines()
mp.add_gridlines(ax, 30)
mp.add_ticks(ax, 30, fontsize=8)
plt.show()
```
